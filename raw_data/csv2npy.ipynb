```python
import pandas as pd
raw_data_directory = 'ODD_Single_muon_10k/'
output_data_directory = '../proccessed_data/'
```

```python
spacepoints_df = pd.read_csv(raw_data_directory + 'spacepoints.csv')
spacepoints_df.drop(spacepoints_df.columns[[-1]], axis=1, inplace=True)
spacepoints_df.drop(spacepoints_df.columns.difference(['event_id', 'x', 'y']), axis=1, inplace=True)

spacepoints_df
```

```python
particles_initial_df = pd.read_csv(raw_data_directory + 'particles_initial.csv')
particles_initial_df.drop(particles_initial_df.columns[[-1]], axis=1, inplace=True)
particles_initial_df.drop(particles_initial_df.columns.difference(['event_id', 'phi', 'pt']), axis=1, inplace=True)

particles_initial_df
```

```python
dataframe = pd.merge(spacepoints_df, particles_initial_df, on='event_id')#.drop('event_id', axis=1)
dataframe
```

```python
import numpy as np
data = dataframe.to_numpy()
data
```

```python
data_filename = output_data_directory + 'x_y_phi_pt.npy'
np.save(data_filename, data)
loaded_data = np.load(data_filename)
loaded_data
```
